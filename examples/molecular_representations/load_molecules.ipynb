# Loading a molecule from a PDB file

This notebook shows how to use the API of `peleffy` to load molecules from an input PDB file.  

## Example 1: PDB with a single chain

This example shows how to load a molecule from a PDB file that contains a single chain corresponding to a hetero molecule. For PDB files with a single chain, the `Molecule` class is used to initializate the molecule. 

### Load required files for this test

```python
from peleffy.utils import get_data_file_path
bnz_pdb_path = get_data_file_path('ligands/benzene.pdb')
```

### Load `peleffy`'s molecule representation 

```python
from peleffy.topology import Molecule

molecule = Molecule(bnz_pdb_path)
display(molecule)
```

## Example 2: PDB with multiple chains

This example shows how to load a hetero molecule from a PDB file of a system. The input PDB file contains the following molecules: 
- Protein
- Hetero molecule (in chain L)

For PDB files with multiple chains, the `PDBFile` class is used to initilizate molecules. Please, note that `peleffy` is only compatible with small molecules. Thus, only hetero molecules will be initialized.   

### Load required files for this test

```python
from peleffy.utils import get_data_file_path
lyz_bnz_path = get_data_file_path('complexes/LYS_BNZ.pdb')
```

### Load all hetero molecules from the PDB 

```python
from peleffy.utils.input import PDBFile

PDBreader = PDBFile(lyz_bnz_path)
molecules = PDBreader.get_hetero_molecules()

for molecule in molecules:
    display(molecule)
```

## Example 3: PDB with multiple hetero molecules
This example shows how to load a molecule from a PDB file of a system. The input PDB file contains the following molecules:

- Protein
- Hetero molecule (in chain L)
- Hetero molecule (in chain G)

### Load required files for this test

```python
from peleffy.utils import get_data_file_path
lig1_lig2_path = get_data_file_path('complexes/LIG1_LIG2.pdb')
```

### Load all hetero molecules from the PDB 

Here, we load all the hetero molecules found in the input file. 

```python
from peleffy.utils.input import PDBFile

PDBreader = PDBFile(lyz_bnz2_path)
molecules = PDBreader.get_hetero_molecules()

for molecule in molecules:
    display(molecule)
```

### Load hetero molecules from a selected chain

We load only the molecules found in a specific chain by setting **selected_chain** to the chain identifier from the PDB file. 

```python
from peleffy.utils.input import PDBFile

PDBreader = PDBFile(lig1_lig2_path)
molecules = PDBreader.get_molecules_from_chain(selected_chain = 'L' )

for molecule in molecules:
    display(molecule)
```
